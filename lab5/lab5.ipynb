## 2.9.1 Создание нового поля через выражение на базе имеющихся колонок

```python
df['Discounted Price'] = df['Mrp'] - df['Sale Price']
df.head()
```

## 2.9.2 Создание новых полей через DataFrame.apply

```python
def calculate_discounted_price(row):
    return row['Mrp'] - row['Sale Price']


df['Discounted Price'] = df.apply(calculate_discounted_price, axis=1)
df.head()
```

## 2.9.3 Создание новых полей через Series.apply

```python
df['Discounted Price'] = df['Mrp'].apply(lambda x: x - 1000)
df.head()
```

## 2.10 Сортировка по Sale Price
